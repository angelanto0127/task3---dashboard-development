```python
import pandas as pd

# Example: Load sample dataset
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)
df.head()
```

```python
from dash import dcc, html, Dash
from dash.dependencies import Input, Output
# from jupyter_dash import JupyterDash # Removed deprecated import
import plotly.express as px

# Create app
# app = JupyterDash(__name__) # Removed deprecated class
app = Dash(__name__) # Used Dash instead

# Layout
app.layout = html.Div([
    html.H1("GDP Over Time by Country"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': i, 'value': i} for i in df['country'].unique()],
        value='India'
    ),
    dcc.Graph(id='line-chart')
])

# Callback
@app.callback(
    Output('line-chart', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_graph(selected_country):
    filtered_df = df[df['country'] == selected_country]
    fig = px.line(filtered_df, x='year', y='gdpPercap', title=f'GDP per Capita: {selected_country}')
    return fig

app.run(mode='inline')
```

## Modify the layout

### Subtask:
Add new interactive components like sliders or radio buttons to the app's layout.


**Reasoning**:
Add a range slider for years and a graph component to the existing layout.



```python
from dash import dcc, html, Dash
from dash.dependencies import Input, Output
import plotly.express as px

# Create app
app = Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("GDP Over Time by Country and Year"),
    html.Div([
        html.Label("Select Country:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': i, 'value': i} for i in df['country'].unique()],
            value='India'
        ),
    ]),
    html.Div([
        html.Label("Select Year Range:"),
        dcc.RangeSlider(
            id='year-slider',
            min=df['year'].min(),
            max=df['year'].max(),
            step=5,
            marks={str(year): str(year) for year in range(df['year'].min(), df['year'].max()+1, 5)},
            value=[df['year'].min(), df['year'].max()]
        ),
    ]),
    dcc.Graph(id='line-chart')
])

@app.callback(
    Output('line-chart', 'figure'),
    [Input('country-dropdown', 'value'),
     Input('year-slider', 'value')]
)
def update_graph(selected_country, selected_year_range):
    filtered_df = df[(df['country'] == selected_country) &
                     (df['year'] >= selected_year_range[0]) &
                     (df['year'] <= selected_year_range[1])]
    fig = px.line(filtered_df, x='year', y='gdpPercap', title=f'GDP per Capita: {selected_country}')
    return fig

app.run(mode='inline')
```
